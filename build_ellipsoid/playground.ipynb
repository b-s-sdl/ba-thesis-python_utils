```python
import numpy as np
import raster_geometry as rg
import matplotlib.pyplot as plt
```

```python
size = 20
sphere = rg.sphere(size,3)
#print(sphere.astype(np.int_))
unique, counts = np.unique(sphere, return_counts=True)
print(unique,counts)

ax = plt.figure().add_subplot(projection='3d')
#print(ellipsoid.nonzero())
x,y,z = sphere.nonzero()
ax.scatter(x, y, z, label='points in (x, z)')
ax.set_xlim(0,size)
ax.set_ylim(0,size)
ax.set_zlim(0,size)
```

```python
size = 20

ellipsoid = rg.ellipsoid(size,[3,8,3])
#print(ellipsoid.astype(np.int_))
unique, counts = np.unique(ellipsoid, return_counts=True)
print(unique,counts)

ax = plt.figure().add_subplot(projection='3d')
#print(ellipsoid.nonzero())
x,y,z = ellipsoid.nonzero()
np.random.seed(8)
xran2 = np.random.choice(len(x), 64, replace=False)
print(xran2)
xran_plot = []
yran_plot = []
zran_plot = []

for i in xran2:
    xran_plot.append(x[i])
    yran_plot.append(y[i])
    zran_plot.append(z[i])

xran_plot = 10 * np.array(xran_plot)-100
yran_plot = 10 * np.array(yran_plot)-100
zran_plot = 10 * np.array(zran_plot)-100

#ax.scatter(x, y, z, label='points in (x, z)')
ax.scatter(xran_plot, yran_plot, zran_plot, label='points in (x, z)')

ax.set_xlim(-size*5,size*5)
ax.set_ylim(-size*5,size*5)
ax.set_zlim(-size*5,size*5)

ax.view_init(elev=20., azim=-35, roll=0)

ax.legend()
```
